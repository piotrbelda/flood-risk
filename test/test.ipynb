```python
from owslib.wfs import WebFeatureService
```

```python
bbox = (20.975,50.001,21.141,50.123)
WFS_URL = "https://wody.isok.gov.pl/wss/INSPIRE/INSPIRE_NZ_HY_MZPMRP_WFS"
wfs = WebFeatureService(url=WFS_URL, version='2.0.0')
response = wfs.getfeature(typename="nz-core:RiskZone", bbox=bbox, maxfeatures=1000)
```

```python
with open("data.xml", "wb") as f:
    f.write(response.read())
```

```python
import geopandas as gpd
```

```python
df = gpd.read_file("data.xml")
```
