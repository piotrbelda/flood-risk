```python
from owslib.wfs import WebFeatureService
```

```python
WFS_URL = "https://wody.isok.gov.pl/wss/INSPIRE/INSPIRE_HY_10k_2019_WFS"
```

```python
bbox = (20.9369, 49.9638, 21.0369, 50.0638)
wfs = WebFeatureService(url=WFS_URL, version='2.0.0')
type_name = "hy-p:Watercourse"
response = wfs.getfeature(typename="hy-p:Watercourse", bbox=bbox, srsname="EPSG:2180")
```

```python
with open("test.xml", "wb") as f:
    f.write(response.read())
```
